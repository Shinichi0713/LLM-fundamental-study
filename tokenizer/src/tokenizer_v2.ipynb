```python
import sentencepiece as spm

# 学習用テキストを準備
text = """私は人工知能が好きです。
あなたはどうですか？
機械学習とディープラーニングは面白いです。
AIは未来を変える技術です。"""

with open("train.txt", "w", encoding="utf-8") as f:
    f.write(text)

# SentencePieceモデルの学習
spm.SentencePieceTrainer.Train(
    input='train.txt',
    model_prefix='sp_demo',
    vocab_size=100,         # 語彙数（サブワード数）
    character_coverage=1.0, # 日本語を全てカバー
    model_type='bpe'        # BPE方式 (unigramにしたいなら'unigram')
)
```

```python
# 学習したモデルを読み込み
sp = spm.SentencePieceProcessor()
sp.load("sp_demo.model")

# テスト文
text = "人工知能は世界を変える"

# トークン化
pieces = sp.encode_as_pieces(text)
ids = sp.encode_as_ids(text)

print("Input:", text)
print("Pieces:", pieces)
print("IDs:", ids)
```

```python
# サブワード → 文字列へ戻す
print(sp.decode_ids(ids))
print(sp.decode_pieces(pieces))
```

```python
samples = ["AIが進化する", "AI進化中"]

for s in samples:
    print(">>", s)
    print(sp.encode_as_pieces(s))
```
