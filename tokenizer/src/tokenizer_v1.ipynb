```python
import MeCab

tagger = MeCab.Tagger("-Ochasen")
text = "私は機械学習と自然言語処理が好きです。"

node = tagger.parseToNode(text)
nouns = []

while node:
    if node.feature.startswith("名詞"):
        nouns.append(node.surface)
    node = node.next

print(nouns)
```

```python
import MeCab
import os

# インストールされたNEologd辞書のパスを検索
# 環境によってパスが異なる場合があるため、以下のコマンドで確認
dic_path = !echo `mecab-config --dicdir`"/mecab-ipadic-neologd"
dic_path = dic_path[0]

# NEologd辞書を指定してTaggerを作成
tagger = MeCab.Tagger(f"-d {dic_path}")

# 解析したいテキスト（例: 最新の言葉を含む）
text_neo = "鬼滅の刃が流行っています。"

# 形態素解析を実行
result_neo = tagger.parse(text_neo)

# 結果の表示
print(result_neo)
```
